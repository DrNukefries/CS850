 2. What are the minimum, maximum, and mean value of each feature?

```python
import pandas as pd
df = pd.read_csv('https://www.statlearning.com/s/Advertising.csv')
print("Min:\n{} \n".format(df.min()))
print("Max:\n{} \n".format(df.max()))
print("Mean:\n{} \n".format(df.mean()))
```

3. Produce a scatterplot matrix of all variables

```python
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix

scatter_matrix(df)
plt.suptitle('Scatterplot Matrix of Advertising Data')
plt.show()
```

 4. Produce a histogram of TV advertising

```python
plt.hist(df['TV'], bins=40)
plt.title('Histogram of TV Advertising')
plt.show()
```
